```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
mall_data=pd.read_csv('/content/Mall_Customers.csv')
print(mall_data)
```

```python
mall_data.isnull().sum()
```

```python
mall_data.shape
```

```python
X = mall_data.iloc[:, [3,4]].values
print(X)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i , init='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('the elbow graph')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
Y = kmeans.fit_predict(X)
```

```python
print(Y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[Y==0,0] , X[Y==0,1] ,s=50, c='green' , label='cluster 1')
plt.scatter(X[Y==1,0] , X[Y==1,1] ,s=50, c='red' , label='cluster 2')
plt.scatter(X[Y==2,0] , X[Y==2,1] ,s=50, c='blue' , label='cluster 3')
plt.scatter(X[Y==3,0] , X[Y==3,1] ,s=50, c='black' , label='cluster 4')
plt.scatter(X[Y==4,0] , X[Y==4,1] ,s=50, c='yellow' , label='cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='cluter')

plt.title('Customers group')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
